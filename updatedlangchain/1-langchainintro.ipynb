### LangChain

LangChain is an open-source framework for developing applications powered by Large Language Models (LLMs). It provides modular components that enable developers to integrate LLMs with external data sources, tools, memory, prompts, and workflows, allowing the creation of intelligent, context-aware, and production-ready AI applications.


### Why was LangChain Created?

A Large Language Model (LLM) by itself can only generate text based on the information it receives.

For example:

User
   │
   ▼
LLM
   │
   ▼
Answer

This works well for simple tasks like:

Writing emails
Explaining concepts
Answering general knowledge questions

However, real-world AI applications require much more.

Suppose a user asks:

"Find the cheapest hotel in Paris and email me the details."

To complete this task, the AI must:

Understand the request.
Search hotel data.
Compare prices.
Choose the cheapest hotel.
Connect to an email service.
Send an email.
Inform the user.

A standalone LLM cannot perform all these tasks.

This is where LangChain comes in.

### Why Do We Need LangChain?

Without LangChain:

User
   │
   ▼
LLM
   │
   ▼
Text Response

The LLM cannot:

Access live data
Use external APIs
Search databases
Remember conversations
Execute Python code
Call tools
Build workflows

With LangChain:

               User
                 │
                 ▼
             LangChain
                 │
 ┌───────────────┼────────────────┐
 │               │                │
 ▼               ▼                ▼

LLM           Tools           Memory

 │               │                │
 └───────────────┼────────────────┘
                 ▼
                 
          Final AI Response

LangChain acts as the orchestrator, coordinating all the components required to build an AI application.

### What Problems Does LangChain Solve?

Without LangChain, developers must manually write code to:

Manage prompts
Connect APIs
Handle memory
Call tools
Parse outputs
Retry failed requests
Switch between LLM providers
Maintain conversation history

LangChain provides reusable components that handle these tasks in a standardized way.

### Output Parsers
Definition

Output parsers convert the raw text generated by an LLM into structured formats such as JSON, Python objects, or Pydantic models.

### Retrieval
Definition

Retrieval is the process of searching external knowledge sources and supplying the retrieved information to the LLM before generating a response.

The pipe operator (|) in LangChain is used to compose multiple Runnable components into a sequential pipeline, where the output of one component automatically becomes the input to the next component.

### Runnable

A Runnable is any LangChain component that accepts an input, performs an operation, and returns an output. Examples include prompt templates, language models, retrievers, tools, and output parsers.

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

response = llm.invoke("Hello")
print(response.content)
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=os.getenv("Gemini_API_KEY")
)

def get_weather(city:str)-> str:
    """Get the weather of the city""" 
    return f"The weather of the {city} is sunny."
agent = create_agent(
    model=llm,
    tools=[get_weather],
    system_prompt="You are a helpful assistant."
)

response = agent.invoke({
    "messages": [
        {
            "role": "user",
            "content": "What is the weather of New York?"
        }
    ]
})

print(response)
```
