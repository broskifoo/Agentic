### Google Gemini Model integration

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI

os.environ["Gemini_API_Key"] = os.getenv("Gemini_API_Key")

model = ChatGoogleGenerativeAI(
     model="gemini-2.5-flash",
    google_api_key=os.getenv("Gemini_API_KEY")

)

response = model.invoke("WHy do parrots talk?")
response.content
```

### Streaming 
streaming is the process of displaying of the results in chunks as they are generated during the process of the generation of the response . the function stream() basically returns an iterator which process the chunks of data generated and displays it and we have to loop it for every chunk created to process it in real time


```python
model.stream("Write a 200 words paragraph on India")
```

```python
for chunk in model.stream("Write a 200 words paragraph on India"):
    print(chunk.text)
```

### Batch

batching is a collection of independent requests to a model can significantly improve performane and reduce costs, as the processign can be done in parallel:

```python
responses = model.batch([
    "Why do parrots fly",
    "Why do airplanes fly",
    "What is quantum computing"
])

for response in responses:
    print(response)
```

```python
model.batch([
    "Why do parrots fly",
    "Why do airplanes fly",
    "What is quantum computing"
],
config={
    'max_concurrency': 5, #Limit to 5 parallel calls
})
```
